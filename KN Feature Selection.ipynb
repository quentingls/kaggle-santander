```python
from pandas import read_csv
```

```python
train_data = read_csv('./data/train.csv')
test_data = read_csv('./data/test.csv')
```

```python
train_data = train_data.set_index('ID')
x_test = test_data.set_index('ID')
```

```python
x_train = train_data.drop(["target"], axis=1)
```

```python
print('train data shape %r, %r'%x_train.shape)
print('test data shape %r, %r'%x_test.shape)
```

```python
from scipy import stats
```

```python
def kn_feature_selection(x_train, x_test, p_threshold):
    dropped_features = []
    for feature in x_train:
        res = stats.ks_2samp(x_train[feature], x_test[feature])
        D, p = res
        print("D : %r,  p : %r"%(D, p))
        if p < p_threshold:
            dropped_features.append(feature)
    
    return x_train.drop(dropped_features, axis=1), x_test.drop(dropped_features, axis=1)
```

```python
x_train, x_test = kn_feature_selection(x_train, x_test, 0.1)
```
